```python
import pygame
```

```python
class station:
    
    def __init__ (self, x=0, y=0, l=20, w=20):
        self.x = x
        self.y = y
        self.lengh = l
        self.width = w
        
    def __str__(self):
        return "({0},{1},{2},{3})".format (self.x, self.y, self.lengh, self.width)
    
    def draw (self, main_area):
        main_area.fill ((0, 150, 0), (self.x, self.y, self.lengh, self.width))
```

```python
class drone:
    
    def __init__ (self, x=0, y=0, l=20, w=20):
        self.x = x
        self.y = y
        self.home_x = self.x
        self.home_y = self.y
        self.lengh = l
        self.width = w
        self.colors = [(0, 0, 150), (150, 150, 150)]
        self.color = (0, 0, 150)
        
    def __str__ (self):
        return "({0},{1},{2},{3})".format (self.x, self.y, self.lengh, self.width)
    
    def draw (self, main_area):
        main_area.fill (self.color, (self.x, self.y, self.lengh, self.width))
    
    def pick_up_parcel (self, parcel_to_pick, main_area):  
        
        if parcel_to_pick.x == parcel_to_pick.destination_x and parcel_to_pick.y == parcel_to_pick.destination_y:
            parcel_to_pick.color = (150, 150, 150)
            #self.return_to_station()     # debugging
            return True   # instead of True      debugging
        
        if self.x == parcel_to_pick.x and self.y == parcel_to_pick.y:
            self.deliver_parcel(parcel_to_pick)
        
        else:
            direction = (self.x - parcel_to_pick.x) / abs(self.x - parcel_to_pick.x) 
            # if direction = 1 then drone needs to shift left on X and up on Y towards the parcel
            # elif direction = -1 then drone needs to shift right on X and down on Y towards the parcel
        
            velocity_x = 1
            difference_x = self.x - parcel_to_pick.x
            updates_number_x = difference_x / velocity_x
            difference_y = self.y - parcel_to_pick.y
            velocity_y = difference_y / updates_number_x 
        
            if self.x != parcel_to_pick.x and self.y != parcel_to_pick.y: 
                self.x = self.x - (velocity_x * direction)
                self.y = self.y - (velocity_y * direction)
            
            #print ((self.x, self.y))
        
    def deliver_parcel (self, parcel_to_deliver): 
        
        if self.x == parcel_to_deliver.destination_x and self.y == parcel_to_deliver.destination_y:
            return True                
        
        else:
            parcel_to_deliver.blinking()
          
            direction = (self.x - parcel_to_deliver.destination_x) / abs(self.x - parcel_to_deliver.destination_x) 
            velocity_x = 4
            difference_x = self.x - parcel_to_deliver.destination_x
            updates_number_x = difference_x / velocity_x
            difference_y = self.y - parcel_to_deliver.destination_y
            velocity_y = difference_y / updates_number_x 
        
            if self.x != parcel_to_deliver.destination_x and self.y != parcel_to_deliver.destination_y: 
                self.x = self.x - (velocity_x * direction)
                self.y = self.y - (velocity_y * direction)  
            if parcel_to_deliver.x != parcel_to_deliver.destination_x and parcel_to_deliver.y != parcel_to_deliver.destination_y: 
                parcel_to_deliver.x = parcel_to_deliver.x - (velocity_x * direction)
                parcel_to_deliver.y = parcel_to_deliver.y - (velocity_y * direction) 
    
    def return_to_station (self):
        
        if self.x == self.home_x and self.y == self.home_y:
            return True
        
        else:
            
            direction = (self.x - self.home_x) / abs(self.x - self.home_x)       
            velocity_x = 1
            difference_x = self.x - self.home_x
            updates_number_x = difference_x / velocity_x
            difference_y = self.y - self.home_y
            velocity_y = difference_y / updates_number_x 
    
            self.x = self.x - (velocity_x * direction)
            self.y = self.y - (velocity_y * direction)        
```

```python
class parcel:
    
    def __init__ (self, x=0, y=0, destination_x=0, destination_y=0):   
        self.x = x
        self.y = y
        self.lengh = 20
        self.width = 20
        self.destination_x = destination_x
        self.destination_y = destination_y
        self.colors = [(0, 0, 150), (150, 150, 150)]
        self.color = (150, 150, 150)
        self.blinking_frequency = pygame.time.Clock()
        
    def __str__ (self):
        return "({0},{1},{2},{3})".format (self.x, self.y, self.lengh, self.width)
    
    def draw (self, main_area):
        main_area.fill (self.color, (self.x, self.y, self.lengh, self.width))
    
    def blinking (self):
        self.color = self.colors[0]
        self.colors.reverse()
        self.blinking_frequency.tick (3)
```

```python
def logistics():
    
    pygame.init()
    main_area = pygame.display.set_mode((1150, 1000))   
    my_clock = pygame.time.Clock()
    
    stations = []
    (pos_x, pos_y) = (60, 50)
    for raw in range (6):
        for column in range (6):
            new_station = station (pos_x, pos_y, 20, 20)
            stations.append (new_station)
            pos_x += 200
        (pos_x, pos_y) = (60, pos_y + 200)
    
    drones = []
    (pos_x, pos_y) = (60, 50)
    for raw in range (3):
        for column in range (3):
            new_drone = drone (pos_x, pos_y, 20, 20)
            drones.append (new_drone)
            pos_x += 400
        (pos_x, pos_y) = (60, pos_y + 400)
        
    parcels_to_pick = []
    parcels_shipped = []
    parcel_1 = parcel (300, 800, 700, 200)      # (300, 900, 400, 800) - works fine
    #parcel_2 = parcel (800, 200, 200, 500)
    parcels_to_pick.append (parcel_1)
    #parcels_to_pick.append (parcel_2)
        
    
         
    while True:
        
        event = pygame.event.poll()
        if event.type == pygame.QUIT:
            break
        if event.type == pygame.KEYDOWN:     
            if event.key == 27:
                break
        
        main_area.fill ((150, 0, 0))
        for each_station in stations:          
            each_station.draw (main_area)
        for each_drone in drones:
            each_drone.draw (main_area)
        for each_parcel in parcels_to_pick:
            each_parcel.draw (main_area)
             
        if parcel_1 in parcels_to_pick:    # check if there are any parcels to pick and find the nearest drone
                            
            parcel_to_pick = parcels_to_pick[0]   
            parcel_location = (parcel_to_pick.x, parcel_to_pick.y)
            distances_to_parcel = []
            for each_drone in drones:
                distance_x = abs (parcel_to_pick.x - each_drone.x)
                distance_y = abs (parcel_to_pick.y - each_drone.y)
                total_distance = distance_x + distance_y
                distances_to_parcel.append (total_distance)
            closest_drone_index = distances_to_parcel.index (min (distances_to_parcel))
            drone_operating = drones [closest_drone_index]   
            
            drone_operating.pick_up_parcel (each_parcel, main_area)
            #drones [drones.index (drone_operating)] = drone_operating   # update the drone's location in drones           
            parcels_shipped.append(each_parcel)   # clones parcel_1 in parcels_shipped each update
        
        if drone_operating.pick_up_parcel(parcel_1, main_area) is True:
            parcel_to_pick.draw(main_area)
            parcels_to_pick = []
            drone_operating.return_to_station()
            
        print ((drone_operating.x, drone_operating.y), (drone_operating.home_x, drone_operating.home_y))    
        my_clock.tick (70)     
        pygame.display.flip()
```
